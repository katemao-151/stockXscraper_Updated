```python
#clean all brand
# change all brand into non capital letter and replace all ' ' with '-'
def changeCapitaltoNoneCapital(brand_list_):
    ret_list = []
    for brand in brand_list_:
        brand = brand.lower()
        if ' ' in brand:
            #print("reached here: "+ brand)
            brand = brand.replace(' ', '-')
            ret_list.append(brand)
        ret_list.append(brand)
    return ret_list
#changeCapitaltoNoneCapital(brand_list)
```

```python
#put brand into API, if total == 0, add the brand to other_list
def getOtherList(cleaned_brand_list_):
    other_list = []
    regular_list = []
    for i in cleaned_brand_list_:  
        sleep(120)
        URL = 'https://stockx.com/api/browse?_tags='+i
        #print(URL)
    #r = requests.get(url = URL)
        send_headers = {
        "User-Agent": headers_list[cleaned_brand_list.index(i)],
        "Connection": "keep-alive",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "Accept-Language": "zh-CN,zh;q=0.8"}
        link = requests.get(URL, headers = send_headers)
        data = link.json()
        if data["Pagination"]["total"] == 0:
            other_list.append(i)
        else:
            regular_list.append(i)
        
    return other_list, regular_list
```
